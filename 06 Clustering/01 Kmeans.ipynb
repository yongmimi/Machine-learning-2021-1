```python
# 군집화(Clustering) - K-Means Algorithm
```

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
```

### Iris 데이터로 군집화

```python
from sklearn.datasets import load_iris
iris = load_iris()
df = pd.DataFrame(iris.data, columns=['sl', 'sw', 'pl', 'pw'])
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

scaler = StandardScaler()
iris_std = scaler.fit_transform(iris.data)

pca = PCA(n_components=2)
iris_pca = pca.fit_transform(iris_std)
```

```python
df['pca_x'] = iris_pca[:, 0]
df['pca_y'] = iris_pca[:, 1]
df.head()
```

```python
### K-Means 모델 생성 / 학습
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3, random_state=2021)
```

```python
kmeans.get_params()
```

```python
kmeans.fit(iris_std)
```

```python
kmeans.labels_
```

```python
df['target'] = iris.target
df['cluster'] = kmeans.labels_
df.tail()
```

```python
result = df.groupby(['target', 'cluster']).count()
result.head()
```

### 시각화 
 - 군집화한 결과를 차원 축소하여 시각화 


```python
markers = ['^','s', 'o']

for i, marker in enumerate(markers):
    x_data = df[df['cluster']==i]['pca_x']
    y_data = df[df['cluster']==i]['pca_y']
    plt.scatter(x_data, y_data, marker=marker)

plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.show()
```

 - 원 데이터를 차원 축소하여 시각화
